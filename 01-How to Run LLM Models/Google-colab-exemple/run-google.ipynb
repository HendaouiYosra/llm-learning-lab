This script starts the Ollama server in the background using threads.
By using a separate thread, the server can be launched without blocking the execution of the main script.
A 5-second delay is added to allow the server to stabilize before performing further actions.



```python
import threading
import subprocess
import time

def run_ollama_serve():
  subprocess.Popen(["ollama", "serve"])

thread = threading.Thread(target=run_ollama_serve)
thread.start()
time.sleep(5)
```
